```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
# For training data (with augmentation)
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# For test / similarity database (NO augmentation)
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
train_dataset = datasets.FashionMNIST(
    root='./data',
    train=True,
    download=True,
    transform=train_transform
)

test_dataset = datasets.FashionMNIST(
    root='./data',
    train=False,
    download=True,
    transform=test_transform
)

print(len(train_dataset), len(test_dataset))
```

```python
image, label = train_dataset[0]

print("Min pixel value:", image.min())
print("Max pixel value:", image.max())
```

```python
plt.imshow(image.squeeze(), cmap="gray")
plt.title(f"Label: {label}")
plt.axis("off")
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False
)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class FeatureCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)

        self.fc = nn.Linear(64 * 7 * 7, 128)  # 128-D embedding

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2)

        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2)

        x = x.view(x.size(0), -1)
        features = self.fc(x)

        return features
```

```python
class CNNClassifier(nn.Module):
    def __init__(self):
        super().__init__()

        self.feature_extractor = FeatureCNN()
        self.classifier = nn.Linear(128, 10)  # 10 classes

    def forward(self, x):
        features = self.feature_extractor(x)
        output = self.classifier(features)
        return output
```

```python
model = CNNClassifier()

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(3):  # 3 epochs are enough
    total_loss = 0

    for images, labels in train_loader:
        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.2f}")
```

```python
model.eval()
feature_model = model.feature_extractor
```

```python
embeddings = []
images_db = []

with torch.no_grad():
    for images, _ in train_loader:
        feats = feature_model(images)
        embeddings.append(feats)
        images_db.append(images)
```

```python
embeddings = torch.cat(embeddings)
images_db = torch.cat(images_db)

print(embeddings.shape)
```

```python
from torch.nn.functional import cosine_similarity
```

```python
def find_similar(query_image, feature_model, embeddings_db, top_k=5):
    """
    query_image: single image tensor [1, 28, 28]
    feature_model: trained CNN feature extractor
    embeddings_db: all embeddings [N, 128]
    top_k: how many similar images to return
    """
    feature_model.eval()

    with torch.no_grad():
        # Get embedding of query image
        query_embedding = feature_model(query_image.unsqueeze(0))  # [1, 128]

        # Compute cosine similarity with all embeddings
        sims = cosine_similarity(query_embedding, embeddings_db)  # [N]

        # Get top K indices
        top_k_indices = sims.topk(top_k).indices

    return top_k_indices
```

```python
# Pick a query image from the dataset
query_image = images_db[100]  # pick any index

# Find top 5 similar images
top_indices = find_similar(query_image, feature_model, embeddings, top_k=5)

print("Top 5 similar indices:", top_indices)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 2))
for i, idx in enumerate(top_indices):
    plt.subplot(1, 5, i+1)
    plt.imshow(images_db[idx].squeeze(), cmap='gray')
    plt.axis('off')
plt.suptitle("Top 5 Similar Clothes")
plt.show()
```

```python
# Step 0: Install / Import
!pip install torch torchvision pillow scikit-learn --quiet

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# Step 1: Preprocessing
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Step 2: Load dataset
train_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=train_transform)
test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=test_transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Step 3: CNN Feature Extractor
class FeatureCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.fc = nn.Linear(64*7*7, 128)
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2)
        x = x.view(x.size(0), -1)
        return self.fc(x)

# Step 4: CNN + classifier
class CNNClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature_extractor = FeatureCNN()
        self.classifier = nn.Linear(128, 10)
    def forward(self, x):
        features = self.feature_extractor(x)
        return self.classifier(features)

# Step 5: Train model
model = CNNClassifier()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(3):  # demo, increase for better results
    total_loss = 0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss:.2f}")

# Step 6: Confusion Matrix
all_preds = []
all_labels = []
model.eval()
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.numpy())
        all_labels.extend(labels.numpy())

cm = confusion_matrix(all_labels, all_preds)
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
plt.figure(figsize=(10,8))
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
plt.title("Fashion-MNIST Confusion Matrix")
plt.show()

# Step 7: Build embeddings database
feature_model = model.feature_extractor
embeddings = []
images_db = []

with torch.no_grad():
    for images, _ in train_loader:
        feats = feature_model(images)
        embeddings.append(feats)
        images_db.append(images)

embeddings = torch.cat(embeddings)
images_db = torch.cat(images_db)

# Step 8: Cosine similarity function
def find_similar(query_tensor, feature_model, embeddings_db, top_k=5):
    with torch.no_grad():
        query_feat = feature_model(query_tensor.unsqueeze(0))
        sims = F.cosine_similarity(query_feat, embeddings_db)
        top_idx = sims.topk(top_k).indices
    return top_idx

# Step 9: Upload your own image
uploaded = files.upload()  # upload jpg/png

query_image_path = list(uploaded.keys())[0]
query_image = Image.open(query_image_path).convert("L")  # grayscale

transform_query = transforms.Compose([
    transforms.Resize((28,28)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
query_tensor = transform_query(query_image)

# Step 10: Find top 5 similar
top_indices = find_similar(query_tensor, feature_model, embeddings, top_k=5)

# Step 11: Show results
plt.figure(figsize=(12,3))
plt.subplot(1,6,1)
plt.imshow(query_image, cmap='gray')
plt.title("Query")
plt.axis('off')

for i, idx in enumerate(top_indices):
    plt.subplot(1,6,i+2)
    plt.imshow(images_db[idx].squeeze(), cmap='gray')
    plt.title(f"Rank {i+1}")
    plt.axis('off')

plt.suptitle("Top 5 Similar Clothes")
plt.show()
```
